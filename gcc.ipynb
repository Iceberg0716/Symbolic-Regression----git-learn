```python
import numpy as np
from scipy.integrate import quad
from scipy.optimize import newton
import matplotlib.pyplot as plt

# 内拉格朗日点函数
def Lagrange(q):
  def equation_x(x, q):
    return (1+q)*x**5 -(2+3*q)*x**4+(1+3*q)*x**3-x**2+2*x-1
  x = newton(equation_x, x0=0.5, args=(q,))
  return x

# 势能函数
def omega(R, zi, theta):
  return np.power(R**2+zi**2, -0.5)+q*np.power(R**2+(1-zi)**2, -0.5)+(1+q)/2*R**2*np.sin(theta)**2+(1+q)/2*zi**2-q*zi

q = 0.01
z0 = Lagrange(q)
omega0 = 1/z0+q/(1-z0)+(1+q)/2*z0**2-q*z0
#外拉格朗日点
z1 = newton(lambda z: omega(0, z, 0) - omega0, -1*z0)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def eggleton_formula(q):
  crbt_q = np.power(q, 1/3)
  return (0.49 * crbt_q**2) / (0.6*crbt_q**2 + np.log(1 + crbt_q))

def f(x):
    return np.piecewise(
        x,
        condlist=[x<=1, x>1],  # 分段条件列表
        funclist=[lambda x: np.power(x,1/3)/np.sqrt(4.545+2.456*np.power(x,2/3)), lambda x: 1/(1.223+1.577*np.power(x,-0.5))]  # 对应表达式
    )

x = np.linspace(0.001,1000, 100)
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8), gridspec_kw={'height_ratios': [2, 1]})
# 主图绘制
ax1.plot(x, f(x), label='$f(x)$', color='#1f77b4', linewidth=2, linestyle='-')
ax1.plot(x, eggleton_formula(x), label='Eggleton Formula', color='#ff7f0e', linewidth=2, linestyle='--')

# 主图美化
ax1.set_xscale('log')  # 对数坐标（因x范围大）
ax1.set_ylabel('Function Value', fontsize=12)
ax1.set_title('Function Comparison', fontsize=14, pad=20)
ax1.legend(framealpha=0.9, shadow=True)
ax1.grid(True, linestyle=':', alpha=0.6)

# 调整边框和标签
for spine in ax1.spines.values():
    spine.set_edgecolor('#dddddd')
ax1.tick_params(axis='both', which='major', labelsize=10)
# 计算相对误差（避免除以零）
relative_error = np.abs(f(x) - eggleton_formula(x)) / np.clip(np.abs(eggleton_formula(x)), 1e-10, None)

# 误差子图
ax2.plot(x, relative_error, color='#d62728', linewidth=1.5, alpha=0.8)
ax2.set_xscale('log')
ax2.set_xlabel('x (log scale)', fontsize=12)
ax2.set_ylabel('Relative Error', fontsize=12)
ax2.grid(True, linestyle=':', alpha=0.4)

# 共享x轴并优化布局
ax1.set_xticklabels([])  # 隐藏主图x标签
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import numpy as np
from scipy.integrate import quad
from scipy.optimize import newton
import matplotlib.pyplot as plt
from scipy.integrate import dblquad
from matplotlib.ticker import PercentFormatter

def Lagrange(q):
  def equation_x(x, q):
    return (1+q)*x**5 -(2+3*q)*x**4+(1+3*q)*x**3-x**2+2*x-1
  x = newton(equation_x, x0=0.5, args=(q,))
  return x


def omega(R, zi, theta):
  return np.power(R**2+zi**2, -0.5)+q*np.power(R**2+(1-zi)**2, -0.5)+(1+q)/2*R**2*np.sin(theta)**2+(1+q)/2*zi**2-q*zi

def f(theta, z):   #最重要的函数
  R0 = newton(lambda r: omega(r, z, 0) - omega0, 0.5)
  R_sol = newton(lambda r: omega(r, z, theta) - omega0, R0)
  return 0.5*R_sol**2


###################################################以上是自制函数


def eggleton(q):
  crbt_q = np.power(q, 1/3)
  return (0.49 * crbt_q**2) / (0.6*crbt_q**2 + np.log(1 + crbt_q))


def g(x):
    result = np.zeros_like(x)
    mask1 = x <= 1
    mask2 = x > 1
    
    # 当 x <= 1 时的表达式
    result[mask1] = -0.11565851 * x[mask1]**(2/3) + 0.49571222 * x[mask1]**(1/3)
    
    # 当 x > 1 时的表达式
    x2 = x[mask2]
    result[mask2] = 0.16671632 + (0.64896418 * (x2 - 0.1)) / (1.75176566 * np.sqrt(x2) + x2)
    
    return result


# 计算相对误差
def relative_error(approx, true):
    return np.abs((approx - true) / true)
```
